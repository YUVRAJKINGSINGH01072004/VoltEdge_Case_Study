```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("Q4_Answer.csv")
df.columns = ['Subcategory', 'Profitability', 'Revenue_Share']
df[['Profitability', 'Revenue_Share']] = df[['Profitability', 'Revenue_Share']].apply(pd.to_numeric)

# Median values for axis splits
x_med = df['Revenue_Share'].median()
y_med = df['Profitability'].median()

# Classify quadrant
df['Quadrant'] = df.apply(
    lambda row: 'Star' if row['Revenue_Share'] >= x_med and row['Profitability'] >= y_med else
                'Question Mark' if row['Revenue_Share'] < x_med and row['Profitability'] >= y_med else
                'Cash Cow' if row['Revenue_Share'] >= x_med else 'Dog',
    axis=1
)

# Color map
colors = {'Star': 'green', 'Cash Cow': 'blue', 'Question Mark': 'orange', 'Dog': 'red'}

# Plot
plt.figure(figsize=(10, 6))
for q in df['Quadrant'].unique():
    sub = df[df['Quadrant'] == q]
    plt.scatter(sub['Revenue_Share'], sub['Profitability'], color=colors[q], label=q, s=100)

for _, row in df.iterrows():
    plt.text(row['Revenue_Share'] + 0.002, row['Profitability'], row['Subcategory'], fontsize=8)

plt.axhline(y_med, color='gray', linestyle='--')
plt.axvline(x_med, color='gray', linestyle='--')
plt.xlabel('Revenue Share (%)')
plt.ylabel('Profitability')
plt.title('Subcategory Positioning')
plt.legend()
plt.tight_layout()
plt.show()
```
